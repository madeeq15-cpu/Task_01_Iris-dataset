# Task 1: Exploring and Visualizing the Iris Dataset

## Objective
Understand how to read, summarize, and visualize the Iris dataset using pandas, matplotlib, and seaborn.

## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2. Load the Dataset

```python
import pandas as pd
import seaborn as sns

iris = sns.load_dataset('iris')
iris.head()
```

## 3. Display Dataset Structure

```python
print("Shape of Dataset:", iris.shape)
print("\nColumn Names:")
print(iris.columns)
print("\nFirst 5 Rows:")
print(iris.head())
```

## 4. Basic Data Summary

```python
iris.info()
iris.describe()
```

## 5. Scatter Plot

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length')
plt.show()
```

## 6. Histogram

```python
plt.figure(figsize=(8,5))
sns.histplot(iris['sepal_length'], bins=20, kde=True)
plt.title('Distribution of Sepal Length')
plt.show()
```

## 7. Box Plot

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=iris, x='species', y='petal_length')
plt.title('Petal Length by Species')
plt.show()
```

# Conclusion

• The Iris dataset contains measurements of three flower species.
• Scatter plots help identify relationships between variables.
• Histograms show how the data is distributed.
• Box plots help detect outliers and understand data spread.
The Iris dataset was successfully explored and visualized using pandas, matplotlib, and seaborn.
